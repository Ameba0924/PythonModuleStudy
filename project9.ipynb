```python
# 9-2 코딩
#random 함수의 suffle(리스트의값을 섞어서 출력)
#random 함수의 ranodm(0부터 1까지의 실수를 발생)
#random 함수의 sample(0,1) 0 리스트 1 갯수 의 난수를 발생

import random as rd 

for i in range(3):
    print(rd.random()) # 0에서 1보다 작은 실수 난수 생성
cards = [[1,'송학'],[2,'메조'],[3,'벚꽃'],[4,'흑싸리'],[5,'초'],[6,'모란']]
print(cards)

for _ in range(3):
    rd.shuffle(cards) # 카드를 랜덤으로 섞음
    print(cards)
    
print(rd.sample(cards,2)) # 카드에서 2개를 랜덤으로 선택 
```

```python
#9-6 코딩

import turtle as t

t.setup(500,400) # 초기 윈도의 크기 조정
t.speed(1) # 1에서 10까지의 거북이 속도 증가, 0이면 초고속
t.pu() # 이동에 선이 그려지지 않도록 설정 penup() , up() 과 같다.
t.goto(0,-100) # 이동

t.pd() # 이동에 선이 그려지도록 설정 pendown(),down()과 같다
t.hideturtle() # 거북이는 보이지 않도록 설정
t.fillcolor('red') # 내부 칠할 색을 설정한다.
t.begin_fill() # 칠하기를 시작한다.
t.circle(100) #크기가 100인 원을 그린다.
t.end_fill() #칠하기를 종료한다.
```

```python
#9-7 코딩 및 다각형그리기

''' 다각형 그리는 함수 정의
   def drawpolygon(n,size)
   한변의 길이가 size인 n각형 그리기

   for i in ragne(n):
       t.forward(size) # 선그리기
       t.left(360/n) # 각도수정 '''

import turtle as t

#색상 리스트
cols = ['red','blue','green','purple','magenta','black',
        'gray','yellow','cyan','orange','aqua']

def drawpolygon(n,size):
    for i in range(n):
        t.pencolor(cols[i % len(cols)]) # 펜 색상을 지정 
        t.forward(size)
        t.left(360/n)

t.setup(500,400) # 초기 윈도우 크기 조정
t.speed(3) #거북이속도3

t.pu() # 이동에 선이 그려지지 않도록
t.goto(-50,-170) #처음 위치로 이동

t.pd() #다각형을 그리기위해 꼬리를 내리고
#한 변의 길이가 100인 다각형 그리기
for i in range(3,12):
    drawpolygon(i,100) # 함수호출 3각형부터 11각형까지 그린다.
```
