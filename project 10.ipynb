```python
# tkinter 내에서의 이벤트 처리
# 이벤트 처리방식 
# 1.이벤트를 발생시키는 버튼이나 메뉴의 키워드인자 'command=함수명' 으로 이벤트핸들러(함수)를 지정하는 방법
# 2.위젯의 함수 bind(이벤트,함수명) 으로 이벤트와 이벤트핸들러인 함수를 연결하는 방법 --> 이방법은 정해진 마우스,키보드의 다양한event처리

# btn = tk.Button(win,text="이벤트처리버튼")
# btn['command'] = 방법1이벤트처리함수명  ==> 커맨드시 def 이벤트처리함수명: 으로 구현 
# btn.bind('<Button-1>',이벤트처리함수명) ==> 바인드시 def 이벤트처리함수명(event): 로 구현해야함 

#Lable 의 인자중 textvariable = 변수명 은 Lable에 문자열을 나타낼 변수를 가지고온다.
#tkinter 는 StringVar 라는 객체를 제공하는데 인자(value='0')로 변수의 초기값을 지정할수있다. -> StringVar.set('0') 과 같은 의미이다.

#10-4코딩 버튼으로 레이블의 정숫값 증가와 감소

from tkinter import *

    def add(): # 1증가시켜 레이블의 문자열에 지정
        n=int(count.get())
        n+=
        count.set(n)
        
    def sub(): # 1감소시켜 레이블의 문자열에 지정
        n=int(count.get())
        n -= 1
        count.set(n)
        
str = ['더하기','빼기']
win = Tk()
#레이블에 표시되며 레이블에 반영될값인 count
#tkinter가 제공하는 StringVar라는 객체에 저장
count = StringVar(value='0') # count.set('0') 또는 value 인자에 첫값인 0을 지정한다.
# 이후 count.get()으로 참조하고  count.set(값)으로 지정

#키워드 인자 textvariable --> 레이블에 표시할 값을 가져올 변수
#키워드 인자 textvariable에 위에서 만든 StringVar 변수인 count를 지정
data = Label(win,width=20,textvariable=count)
data.grid(row=0,column=0,columnspan=2)

badd = Button(win,text=str[0],command=add) # 버튼을 누르면 add함수를 호출한다.
bsub = Button(win,text=str[1],command=sub) # 버튼을 누르면 sub함수를 호출한다.
badd.grid(row=1,column=0)
bsub.grid(row=1,column=1)

win.mainloop()
```

```python
#canvas 선,다각형,원등을 그리기 위한 위젯으로 활용
#캔버스의 키워드인자 relief는 캔버스의 테두리 모양으로 flat,groove,raised,ridge,solid,sunken 등을 지정할수있다. ex) relief='solid' 
#외곽 테두리 두깨(borderwidth) 는 bd에 지정할수있다.
#함수 pack()에서 확장여부를 expand=True 로 지정할수있으며, 윈도크기에 변화에 따라 채워질 방향을 가로 세로 양쪽인 fill='both' 로 지정!


#!!!! 캔버스에서 왼쪽 마우스 클릭 이벤트는 <Button-1> 이며 , 처리함수를 연결하기위해서
#canvas.bind(<Button-1>,click) 이런식으로 한다.
#!!!! 캔버스에서 왼쪽 마우스 릴리즈 이벤트는 <ButtonRelease-1> 이며, 처리함수를 연결하기위해서
#canvas.bind(<ButtonRelease-1>,release) 로 구현한다.

# 10 - 7 코딩 캔버스에 직선 그리기

from tkinter import *

win = Tk()
win.title('라인그리기')
win.geometry('640x400+100+100') # 좌표설정 x100 y100 설정한것, 

def click(event): # bind와 연결해 사용하려면 인자에 event를 꼭 넣어주어야함 
    global sX,sY
    print('클릭위치' ,event.x, event.y)
    sX,sY = event.x,event.y

def release(event):
    global eX,eY
    print('릴리즈 위치', event.x,event.y) 
    eX,eY = event.x,event.y
    #직선 라인 그리기
    canvas.create_line(sX,sY,eX,eY,fill='red',width=2) #캔버스의 crete_line() 함수는 직선을 그려준다 
    #인자로는 (클릭위치x,클릭위치y,릴리즈위치x,릴리즈위치y 까지 , fill색상으로 , 직선두깨width 로 그려줌)
    
canvas = Canvas(win,relief='solid',bd=2) #캔버스객체 생성 (win위에그리겠다, relief로 테두리모양지정 , bd로 외곽테두리두께지정(borderwidth))
canvas.pack(expand=True,fill='both') #pack을통해서 expand=True확장하고 , 윈도크기변화에따라 채워질방향을 가로세로양쪽 fill='both' 로지정
#여기서 expand는 요구되지않은 다른공간을 사용하겠다는 의미이고 #fill은 요구되었지만 사용되지않은 공간만큼 늘릴수있도록한다. 즉, 창너비에맞춰 다쓰지않아도 늘리고 사용할수있도록함 
#이설명이어렵다면 https://m.blog.naver.com/PostView.nhn?blogId=sisosw&logNo=221411972022&proxyReferer=https%3A%2F%2Fwww.google.com%2F 를참고하도록한다.
#왼쪽 마우스 버튼 클릭 바인딩
canvas.bind("<Button-1>",click) # "<Button-1>" 왼쪽마우스클릭이벤트에 click함수를 발생 
canvas.bind("<ButtonRelease-1>",release) #"<ButtonRelease-1>" 왼쪽마우스클릭했다가떼는 이벤트에 release함수를 발생

win.mainloop() 
```

```python
#10-10 파이게임이용하기 

import pygame

# 색상 정의
WHITE = (255,255,255) # RGB 값 0,0,0 은 BLACK

#윈도우 크기 정의

size = (300,200)

#파이 게임 초기화
pygame.init()

#화면 크기 지정해 스크린을 생성
screen = pygame.display.set_mode(size) #display.set_mode() 함수를 통해 인자인 크기를이용해 생성

#제목 (캡션) 지정
pygame.display.set_caption('파이게임윈도!') # display.set_caption() 함수를통해 제목 지정

#윈도 종료 플래그로 사용되는 변수 초기값 지정
done = False
#메인 루프
while not done:
    for event in pygame.event.get(): #여러 이벤트를 받아서 처리한다는것.
        if event.type == pygame.QUIT: # 이벤트타입이 윈도우 종료키이면
            done = True #프로그램을 지정하기위해 플래그값을 true로 변경시켜줌

        screen.fill(WHITE) # 스크린색상을 지정할수있다.
        pygame.display.update() # 화면의 필요한 부분만을 수정 ==> 화면에 그림을 그린 후에는 반드시 화면을 update처리해주어야함

pygame.quit() # 메인루프를 빠져나와 파이게임을 종료시킨다.
```

```python
# 10-12 코딩 반복 눈내리기코딩 

import pygame,random

# 게임 엔진 초기화
pygame.init()

BLACK=[0,0,0]
WHITE=[255,255,255]
SIZE=[300,300] # 윈도우 크기 지정
SNOW_CNT = 70 # 눈의 개수 지정

screen = pygame.display.set_mode(SIZE)
pygame.display.set_caption('눈내리기')

#눈의 좌표 리스트 

snow_list = []

# snow_list 에  SNOW_CNT 수만큼의 좌표 저장

for i in range(SNOW_CNT):
    x = random.randrange(0,SIZE[0]) # x 좌표 저장 
    y = random.randrange(0,SIZE[1]) # y 좌표 저장
    snow_list.append([x,y]) # 리스트에 추가 
    
#화면 수정에 사용될 시계 저장
clock = pygame.time.Clock()

#메인 루프

flag = False 
while not flag:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            flag = True  # 무한루프 종료
            
            
    #배경색 지정
    screen.fill(BLACK)
    
    #눈 내리는 모습 그리기
    for i in range(len(snow_list)):
        
        #눈모양의 원 그리기
        radius = 1  # <<<< 눈의 크기
        # radius = random.randint(1,3)
        pygame.draw.circle(screen,WHITE,snow_list[i],radius) #화면값 , 색상 , 좌표값 , 눈크기 지정
        
        #눈 snow_list[i] 의 y좌표를 1증가시켜서 눈이 내리도록 수정
        snow_list[i][1] += 1
        # snow_list[i][1] += random.randint(1,3)
        
        #눈 snow_list[i]의 x 좌표를 (-1,0,1) 중에 하나에서 선택 이전값이 더해 대입 
        snow_list[i][0] += random.randint(-1,1)
        
        # snow_list[i] 가 바닥으로 내려가면, 즉 보이지 않게 되면
        if snow_list[i][1] > SIZE[1]: # snow_list[i][1]은 눈의 y좌표이다. SIZE[1]은 윈도우의 세로크기
            #y좌표를 다시 음수로 설정해줌
            snow_list[i][1] = random.randint(-5,0)
            #x좌표도 다시 지정해줌 ,  snow_list[0]은 눈의 x좌표이고 SIZE[0]은 윈도우의 가로크기
            snow_list[i][0] = random.randint(0,SIZE[0])
            
    pygame.display.update() # 화면 업데이트
    #초당 수정될 프레임 수 지정,  초당 20 프레임 화면이 수정됨 
    clock.tick(20)


pygame.quit() #무한루프를 빠져나오면 파이게임 종료 
```
